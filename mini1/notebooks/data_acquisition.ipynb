```python
import pandas as pd
import requests
```

```python
url = "https://data.cdc.gov/resource/i46a-9kgh.json"
params = {
    "$select": "stateabbr,countyname,countyfips,BINGE_AdjPrev",
    "stateabbr": "CA"
}

r = requests.get(url, params = params)
r.raise_for_status()

binge_ca = pd.DataFrame(r.json())
```

```python

binge_ca.rename(columns={'countyname': 'County'}, inplace=True)
binge_ca.head()
```

```python
income_path = "../data/raw/2023-income-limits.csv"

income_df = pd.read_csv(income_path)
income_df.head(5)
```

```python
df = pd.merge(income_df, binge_ca, on='County')
df = df.loc[:, ['County', 'stateabbr', 'countyfips', 'BINGE_AdjPrev', 'AMI']]
df.head()
```

```python
print(df.shape)
```

```python
output_file = '../data/interim/combined_data.csv'
df.to_csv(
    output_file,
    index=False,
    encoding="utf-8"
)

print(f"Output file: {output_file}")
```

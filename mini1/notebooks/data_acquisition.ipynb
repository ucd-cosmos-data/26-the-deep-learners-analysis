```python
import pandas as pd
import requests
```

```python
url = "https://data.cdc.gov/resource/i46a-9kgh.json"
params = {
    "$select": "stateabbr,countyname,countyfips,BINGE_AdjPrev",
    "stateabbr": "CA"
}

r = requests.get(url, params = params)
r.raise_for_status()

binge_ca = pd.DataFrame(r.json())
```

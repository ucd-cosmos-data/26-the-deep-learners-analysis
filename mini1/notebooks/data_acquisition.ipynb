```python
import pandas as pd
import requests
```

```python
url = "https://data.cdc.gov/resource/i46a-9kgh.json"
params = {
    "$select": "stateabbr,countyname,countyfips,BINGE_AdjPrev",
    "stateabbr": "CA"
}

r = requests.get(url, params = params)
r.raise_for_status()

binge_ca = pd.DataFrame(r.json())
```

```python

binge_ca.rename(columns={'countyname': 'County'}, inplace=True)
binge_ca.head()
```

```python
income_path = "../data/raw/us_income.csv"

income_df = pd.read_csv(income_path)
display(income_df.head(5))

cali_df = income_df[income_df['State'] == 'California']
display(cali_df.head(5))

cali_df['County'] = cali_df['County'].apply(lambda x: x[:len(x)-7])

cali_df['FIPS'] = cali_df['FIPS'].apply(lambda x: "0" + str(x))

print(cali_df.shape)
```

```python
display(cali_df.head(5))
binge_ca.head()
```

```python
df = pd.merge(cali_df, binge_ca, on='County')
display(df.head())
df = df.loc[:, ['County', 'stateabbr', 'countyfips', 'BINGE_AdjPrev', 'Median Household Income']]
df.head()
```

```python
output_file = '../data/interim/combined_data.csv'
df.to_csv(
    output_file,
    index=False,
    encoding="utf-8"
)

print(f"Output file: {output_file}")
```
